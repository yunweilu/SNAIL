```python
import numpy as np
import matplotlib.pyplot as plt

gamma =1/2e4
A_values = np.linspace(0.1e-3, 5e-3, 20) * 2 * np.pi

results = [gamma*(((2*A/(0.1*2*np.pi))**(2/3)/4)**2 + 0.1**4*((0.1*2*np.pi/(A))**(2/3))/2) for A in A_values]

plt.figure(figsize=(10, 6))
plt.plot(A_values/(2*np.pi), results, 'b-o', markersize=4)
plt.xlabel('A (in units of 2Ï€)')
plt.ylabel('Result')
plt.title('Expression evaluation vs A')
plt.grid(True, alpha=0.3)
plt.show()
```
