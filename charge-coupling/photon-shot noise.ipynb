```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
T = 90*1e-6
gamma = 1/T
# n_bar = 1/np.sqrt(23.5e-3/(T/2))
n_bar = 1/(np.exp(6.5/(0.05*20.836))-1)
# Chi values
chi = np.linspace(gamma*0.01, gamma*100, 1000)
# chi = gamma
ratio = chi / gamma

# Factor calculation
factor = np.real(np.sqrt(((1 + 1j*ratio)**2 + 4*1j*ratio*n_bar)) - 1) / 2

# a calculation
a = 1 / (gamma * factor)*1e3

# Plot
plt.figure(figsize=(10, 6))
plt.yscale('log')
plt.xscale('log')
plt.plot(chi/1e3, a, label='$a$ vs. $\chi$')
plt.xlabel('$\chi$ (KHz)')
plt.ylabel('Dephasing time due to photon shot noise (ms)')
plt.grid(True)
plt.show()
```
