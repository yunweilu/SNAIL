```python
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
import matplotlib.pyplot as plt
import numpy as np
from qutip import *
import scipy.special as sp
from tqdm import tqdm
import scipy
```

```python
qutip.__version__
```

```python
delta = 0.  * 2*np.pi
eps0  = 1. * 2*np.pi
w_a = np.sqrt(delta**2 + eps0**2)

A     = 0.0* 2*np.pi

omega = (1)*2*np.pi
T     = 2*np.pi / omega

tlist  = np.linspace(0.0, 30 * T, 1001)
psi0   = np.sqrt(1/2)*(qutip.basis(2,1)+qutip.basis(2,0))
```

$$
H(t) = -\frac{\delta}{2}\sigma_{x} - \frac{\epsilon_{0}}{2}\sigma_{z} + \left(\frac{A}{2}\sigma_{x}\right)\cos(\omega t)
$$


```python
H0 = - delta/2.0 * qutip.sigmax() - eps0*qutip.Qobj(np.array([[0, 0], [0, 1]]))
H1 = A/2 * qutip.sigmax()
H = [H0, [H1, lambda t: np.cos(omega * t)]]
```

```python
# noise power spectrum
gamma1 = 0.01
def noise_spectrum(w):
    return   gamma1* (w==0)
    # return   gamma1 * np.abs(w)**2 * (w>0)
```

```python
temperature = 0e-3 # unit K 
h = 6.626e-34 
kB = 1.38e-23 
w_th = temperature * (kB / h) * 2 * np.pi * 1e-9    
```

```python
opt = qutip.Qobj(np.array([[0, 0], [0, 1]]))
```

```python
output = qutip.fmmesolve(
    H, psi0, tlist, [opt],
    spectra_cb=[noise_spectrum], T=T, w_th=w_th,
     options={"store_floquet_states": True}
)
```

```python
from auxiliary_func import *
A,w = obtain_Aw(H, [opt], [noise_spectrum], T, w_th)
print(w.to_array())
print(A.to_array())
```

```python
p_x = np.zeros(tlist.shape, dtype=np.complex128)
p_y = np.zeros(tlist.shape, dtype=np.complex128)
p_z = np.zeros(tlist.shape, dtype=np.complex128)
for idx, t in enumerate(tlist):
    psi_t = output.floquet_states[idx]
    p_x[idx] = qutip.expect(qutip.sigmax(), psi_t)
    p_y[idx] = qutip.expect(qutip.sigmay(), psi_t)
    p_z[idx] = qutip.expect(qutip.sigmaz(), psi_t)
```

```python
plt.plot(tlist, np.real(p_x), 'r-')
plt.plot(tlist, np.real(p_y), 'b--')
plt.plot(tlist, np.real(p_z), 'g')
plt.xlabel('Time')
plt.ylabel('<sigma>')
plt.grid()
plt.show()
```

```python
from scipy.optimize import curve_fit
```

```python
def model_function(t, a, c, tau):
    return a*np.exp(-t / tau) + c
```

```python
popt_x, pcov = curve_fit(model_function, tlist, np.real(p_x), p0=[1, 1, 1], maxfev=50000)
a_fit, c_fit, tau_fit_z = popt_x
print(f"Fitted parameters: a = {a_fit:.4f}, c = {c_fit:.4f}, tau = {tau_fit_z:.4f}")
```
